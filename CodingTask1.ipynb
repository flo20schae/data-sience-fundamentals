CT-1.1

```python
def remove_header(file_path, header_path):
    # Open the original file and read all the lines
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Find the index of the first line that does not start with '#'
    header_end = 0
    for i, line in enumerate(lines):
        if not line.startswith('#'):
            header_end = i
            break

    # Extract the header information and save it to a separate file
    header = lines[:header_end]
    with open(header_path, 'w') as header_file:
        header_file.writelines(header)

    # Remove the header lines from the original file and save the remaining data
    data = lines[header_end:]
    with open(file_path, 'w') as file:
        file.writelines(data)

remove_header(file_path, header_path)
with open("CCD-Header.txt", 'r') as h:
  for line in h:
    print(line)
```

```python

def plot_difference_improved(df,wav1, wav2):
    diff = pd.DataFrame({'Timestamp' : df['Unnamed: 1'].copy(),
                         'Difference' : df[wav1] - df[wav2],
                         '{}'.format(wav1): df[wav1],
                         '{}'.format(wav2): df[wav2]})
    diff['Timestamp'] = (diff['Timestamp'] - diff['Timestamp'].iloc[0])/1000
    plot = diff.plot(x='Timestamp', y=['Difference', '{}'.format(wav1), '{}'.format(wav2)],
                     xlabel= 'Time [sec]', ylabel='Absorbance [AU]', 
                     title='Difference in Absorbance at {}nm and {}nm'.format(wav1, wav2))
    plot.get_figure().savefig('plot.jpg')
     
```
